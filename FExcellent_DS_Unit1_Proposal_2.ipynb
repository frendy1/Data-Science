<h1>
    <span>A/B TESTING PROPOSAL
    </span>
</h1>
<p class=MsoNormal>
    <span>Frendy Excellent, Thinkful, 4/13/2019</span>
</p>


<h3>
    <span>Fictional Situation</span>
</h3>

<p class=MsoNormal>
    <span>AutoZone wishes to deploy a website to increase brand awareness and customer traffic. A majority of senior leadership has adopted a resolution to include the term “auto zone” in the domain name of its website. However, a small group of internal marketing expert has been pushing for the inclusion of the term “auto part” somewhere in the domain name of the company’s website. As a result, leadership has brought in a consulting firm to help clarify the impact of not including the term “auto part” in the organization’s domain name.
    </span>
</p>

<h3>
    <span>Problem Statement</span>
</h3>

<p class=MsoNormal>
    <span>Leadership needs to validate that using the term &quot;auto zone&quot; or &quot;auto part&quot; in the domain name of the company does not make a difference, concerning brand awareness and customer traffic.
    </span>
</p>

<h3>
    <span>Model</span>
</h3>

<p class=MsoNormal>
    <span>Customers use a search engine to find auto parts stores in their neighborhood. Customers submit search terms (such as &quot;auto&quot; and &quot;part&quot;) to a search engine. As a result of customers’ queries, the search engine returns several pages of rankings of auto parts vendors (ranked from 1 to n).  Empirical research (see reference 3) has revealed that about 94 percent of customers click (when they choose to do so, instead of just viewing results) on the top two domain names returned by the search engine (rank 1 and 2). Note: Only organic search results are considered in the model (the model ignores &quot;ads&quot; returned on top of the ranked results’ web page).
    </span>
</p>


<h3>
    <span>Metric</span>
</h3>

<p class=MsoNormal>
    <span>To operationalize brand awareness and customer traffic measurements, and based on the above model, customer click through rate (count of clicks on link divided by count of views of link on search result page) has been adopted as the metric of choice in this investigation.
    </span>
</p>


<h3>
    <span>Null Hypothesis
    </span>
</h3>

<p class=MsoNormal>
    <span>For queries with at least both terms &quot;auto &quot;and &quot;part&quot; in submitted searches, it can be stated, with 95 percent confidence, that the click through rate of a domain name that includes the term &quot;autozone&quot; is not significantly different from the click through rate of a domain name that includes the &quot;autopart&quot; term.
    </span>
</p>

<h3>
    <span>Alternative Hypothesis
    </span>
</h3>

<p class=MsoNormal>
    <span>For queries with at least both terms &quot;auto&quot; and &quot;part&quot; in submitted searches, it can be stated, with 95 percent confidence, that the click through rate of a domain name that includes the term &quot;autozone&quot; is significantly different from the click through rate of a domain name that includes the &quot;autopart&quot; term.
    </span>
</p>


<h3>
    <span>
        Experimental Design
    </span>
</h3>

<p class=MsoNormal>
    <span>The A/B testing methodology will be applied in this investigation. In brief, domain names that include the terms &quot;auto&quot; and &quot;part&quot; will be the control. The company’s domain name that includes the terms &quot;auto&quot; and &quot;zone&quot; will be the variable being evaluated.
    </span>
</p>


<h3>
    <span>
        Sample Size and Statistical Power
    </span>
</h3>

<p class=MsoNormal>
    <span>The experiment requires a sample size that will allow a maximum of 5 percent chance of erroneously rejecting the null hypothesis (Significance Level). In addition, the experiment needs a sample size that will allow a 20 percent chance of erroneously accepting the null hypothesis, with 80 percent chance of accepting the alternative hypothesis (Statistical Power). (See references 1, 2, 4, and 5)
    </span>
</p>


<h3>
    <span>
       Testing Environment, Data Collection and Measurement
    </span>
</h3>

<p class=MsoNormal>
    <span>The experiment does not use any tools to directly collect and measure customer click through rates for the control and target. Instead, in light of the above sample size and power requirements, the consultants will purchase three-month’s (01 March, 2006 - 31 May, 2006) worth of search data (36,389,567 lines of data) from the search engine firm, AOL (see reference 3). Next, using the purchased dataset, the investigators will measure customer click through rates for the control and target.
    </span>
</p>

<h3>
    <span>
       Analysis and Discussion
    </span>
</h3>

<p class=MsoNormal>
    <span>A two-tailed t-test will be performed and discussed here.  The two-tailed t-test helps assess the difference in click through rates for control and target. A one-tailed t-test, which will not be performed here, helps decide whether to replace “autozone” with “autopart” in the domain name. In addition, weaknesses and bias will be discussed here. </span></p>
