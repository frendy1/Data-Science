```python

from scipy.stats import ttest_ind, ttest_ind_from_stats
```

```python
abar = 5
bbar = 8

astd = 1
bstd = 3

na = 200
nb = 500

t2, p2 = ttest_ind_from_stats(abar, astd, na,
                              bbar, bstd, nb,
                              equal_var=False)
print("ttest_ind_from_stats: t = %g  p = %g" % (t2, p2))
```

```python
abar = 1090
bbar = 999

astd = 400
bstd = 30

na = 900
nb = 100

t2, p2 = ttest_ind_from_stats(abar, astd, na,
                              bbar, bstd, nb,
                              equal_var=False)
print("ttest_ind_from_stats: t = %g  p = %g" % (t2, p2))
```

```python
abar = 45
bbar = 40

astd = 45
bstd = 40

na = 2000
nb = 2000

t2, p2 = ttest_ind_from_stats(abar, astd, na,
                              bbar, bstd, nb,
                              equal_var=False)
print("ttest_ind_from_stats: t = %g  p = %g" % (t2, p2))
```
