        Problem: Will new uniforms help a gym's business?
Answer: Control: uniform color A, Variable: new color B.
Select two groups of randomly selected gym locations: group X and group Y. Assign existing color A to group X, assign color B to group Y. Collect a random sample, until sign-up count is large enough, based on confidence level and statistical power estimates.
Null: There is no significant difference between colors.
Alternative: There is a significant difference between colors.
Perhaps, new color improves signing up new customers.
Measure error rate in data entry as volume of sign-ups increases. 
    </span>

        Problem: Will a new homepage improve my online exotic pet rental business?
Answer: Control: existing home page, variable: new home page
Serve control, at random, odd IP address. Serve variable, at random, even IP address. 
Null Hypopthesis: There is no significant difference between the home page versions.
Alternative Hypopthesis: There is a significant difference between the home page versions.
New home page version performs better than existing version.
Observe the most frequent location that visitors click on the new page.

Answer: Control: Email without 'please read' in the subject, variable: Email with 'please read' in the subject
Split mailing list into two groups (50:50) of random recipients. Send control email to one random group of recipients from mailing list. Send variable email to other random group of recipients from mailing list.
Null Hypopthesis: There is no significant difference between the versions of email subjects.
Alternative Hypopthesis: There is a significant difference between the versions of email subjects.
