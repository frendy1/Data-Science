<h1>
    <span>A Quick Exploration of 16<sup>th</sup> Hole Putts at the Waste Management Phoenix Open
    </span>
</h1>
<p class=MsoNormal>
    <span>Frendy Excellent, Thinkful, 4/1/2019</span>
</p>



<p class=MsoNormal>
    <span>
        <img width=474 height=283 src="FExcellent_DS_Fund_Capstone3_files/image707.gif">
    </span>
</p>

<p class=MsoNormal>
    <span>For a specific hole of a particular
golf course, does the golf ball have a tendency to settle at a certain location
or locations on the green, as a result of a putt? 
    </span>
    <img width=290 height=151 src="FExcellent_DS_Fund_Capstone3_files/image607.gif" align=left hspace=12>
    <span>To answer this question,
the investigation targeted two types of putts: putts driven into the 16<sup>th</sup>
hole and putts missed (e.g., putts that overshot into the hole). A PGA ShotLink
dataset for the 16<sup>th</sup> hole at the Waste Management Phoenix Open was analyzed.
The ShotLink dataset includes GPS-like coordinates to help pinpoint the golf
ball locations on the course. The 16<sup>th</sup> hole dataset investigation targeted all
16<sup>th</sup> hole green putts (for all rounds played) in year 2015, 2016, and 2017
Waste Management Phoenix Open. Note that even the elevation at sea level is
present in the ShotLink dataset. This could be an important piece of data for
future putt investigations, judging from the intimidating geography of the 16<sup>th</sup>
hole (and high-decibel crowd). Mounds and slopes have been ignored here for the
sake of simplicity. However, they probably play a significant role in the
match. Nonetheless, since a picture is worth more than words, we will look a
few charts to digest potentials sources of putting effects below.
    </span>
</p>

<h3>
    <span>Holed Out Putts</span>
</h3>

<p class=MsoNormal>
    <span>The 2015,
2016, and 2017 Waste Management Phoenix Open competition resulted in 1,211
putts making it into the 16<sup>th</sup> hole (holed out putts). Furthermore, those putts
consist of 595 far holed out putts and 616 near holed out putts, for a total
1211 such putts. The putts that made it into the hole exhibit two clusters. The
top left corner of the following chart represent the putts that are farther
from the hole (&quot;far holed out putts&quot;. The bottom right corner of the
chart represent the putts that are closer the hole (&quot;near holed out
putts&quot;).
    </span>
</p>

color:black;font-weight:normal'>In brief, a quick look at the far holed out
putts yielded 99 far holed out putts in the first cluster, 179 far holed out

121 far holed out putts in the fourth cluster 1, for a total of 531.

<h3>
    <span>
        What have we found so far?
    </span>
</h3>

<p class=MsoNormal>
    <span>The
above investigation indicates that holed out putts display patterns as to green
location, while missed putts show no apparent location preference. So, the data has
revealed, thus far, eight locations on the green were successful putts have a
tendency to settle, while two additional locations on the green result in
missed putts. What is needed is a bird’s eye view of the action on the green.
&nbsp;To gain valuable insights (using a “view from above”), a polar viewpoint
of the data was adopted, as shown below. 
    </span>
</p>

<p class=MsoNormal>
    <span>
        <img width=601 height=601 id="Diagram 38" src="FExcellent_DS_Fund_Capstone3_files/image007.gif">

<p class=MsoNormal>
    <span>It appears that some golfers who placed the ball into the
16<sup>th</sup> hole had a tendency to land their putts on the right side of
the 16<sup>th</sup> hole and farther as well. Several unsuccessful distant
putts can be seen on the right side of the hole. Thus, it is probably fair to
propose that the right-sided putts are activities resulting from struggling
golfers. Putts not landing on the right side of the hole are perhaps made by
seasoned golfers, whom are probably not deviating much from the fairway.
Furthermore, common sense dictates that closer putts are more likely to be
successful, as well distant putts launched by seasoned golfers.</span></p>

<p class=MsoNormal>

    <span>As you are
aware, we must inspect the following three charts to figure out the real story
here (otherwise, the experts will remind us!). The histograms imply that the
long and short putts that end up into the hole are mostly from two feet, while
most of the missed putts are 20 feet way. Next, it is hard to predict missed
putts beyond 20 feet, as shown by the probability plot for missed putts.
Finally, the boxplots (last chart) yields the same conclusion as the
histograms.

<h3>
    <span>Conclusion</span>
</h3>

<p class=MsoNormal>
    <span>This quick
review seems to tell us, in part, that it is good practice to place ball at
about 2two feet from the whole. In contrast, a putt from 20 feet or more will
more likely have to be putted again. Further investigation of how putts that
went beyond the hole fared would be an interesting topic for further
investigation.
